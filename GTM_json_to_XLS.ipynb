```python
account = ""
startRow = 0
df_columns = ['S.No', 'Tag Name', 'Tag Type', 'Tag Configuration', 'Trigger', 'Trigger Configuration']
df_dict_empty = {'S.No': "", 'Tag Name': "", 'Tag Type': "", 'Tag Configuration': "", 'Trigger': "", 'Trigger Configuration': ""}

result = {}
for key, value in account_id_name.items():
    result[key] = pd.DataFrame(columns=df_columns)

def main(filename):
  global startRow
  global account
  global main_df
  global result

  json_data = json.loads(open('{0}'.format(filename)).read())

  triggers = json_data['containerVersion']['trigger'] if('trigger' in json_data['containerVersion']) else {}
  tags = json_data['containerVersion']['tag'] if('tag' in json_data['containerVersion']) else {}

  account_id = json_data['containerVersion']['accountId']
  account = account_id_name[account_id]
  main_df = pd.DataFrame(columns=df_columns)

  df_dict = {'S.No': "", 'Tag Name': "", 'Tag Type': "", 'Tag Configuration': "", 'Trigger': "", 'Trigger Configuration': ""}
  df = pd.DataFrame(columns=df_columns)
  container_name = json_data['containerVersion']['container']['name']
  container_id = json_data['containerVersion']['container']['publicId']
  container_type = json_data['containerVersion']['container']['usageContext'][0]

  def getTriggerParamsCustom(trigger_parameters):
    trigger_parameter_data = ""
    for trigger_parameter in trigger_parameters:
      trigger_parameter_data += "{0} - {1} - {2}\n".format(trigger_parameter['parameter'][0]['value'], trigger_parameter['type'], trigger_parameter['parameter'][1]['value'])
    return trigger_parameter_data

  def getTriggerNameViaId(id):
    if("2147479553" == id): 
      return "All Pages"
    for y in triggers:
      if(y['triggerId'] in id):
        return y['name']

  def getTriggerViaId(id):
    if("2147479553" == id): 
      return id
    for y in triggers:
      if(id in y['triggerId']):
        return y

  def getTriggerParams(y):
    trigger_parameter_data = ""
    if(y == "2147479553"):
      return "All Pages"
    if("parameter" in y):
      for trigger_parameter in y["parameter"]:
        if(trigger_parameter['type'] != "LIST"):
          trigger_parameter_data += "{0} - {1}\n".format(trigger_parameter['key'], trigger_parameter['value'])
        elif(trigger_parameter['type'] == "LIST" and y['type'] == "TRIGGER_GROUP"):
          trigger_parameter_data = "\nTrigger Group"
          for count, list_items in enumerate(trigger_parameter["list"]):
            trigger_parameter_data += "\n{0}: TRIGGER - {1}\n".format(str(count+1), getTriggerNameViaId(list_items['value']))
            trigger_parameter_data += getTriggerParams(getTriggerViaId(list_items['value']))
          trigger_parameter_data += "\n\nThis trigger group fires on\n"
    if("filter" in y):
      trigger_parameter_data += getTriggerParamsCustom(y["filter"])
    if("customEventFilter" in y):
      trigger_parameter_data += getTriggerParamsCustom(y["customEventFilter"])

    return trigger_parameter_data

  def getTriggers(firing_trigger_id):
    if("2147479553" in firing_trigger_id): 
      df_dict["Trigger"] += "\nTrigger Name: {0}\nTrigger Type: {1}\n".format("All Pages", "Built-In")
      df_dict["Trigger Configuration"] += "All Pages"
    for y in triggers:
      trigger_id = y['triggerId']
      trigger_type = y['type']
      trigger_name = y['name']
      if(trigger_id in firing_trigger_id): 
        df_dict["Trigger"] += "\nTrigger Name: {0}\nTrigger Type: {1}\n".format(trigger_name, trigger_type)
        trigger_parameter_data = "\n\nTrigger: {0}\n".format(trigger_name)
        trigger_parameter_data += getTriggerParams(y)
        df_dict["Trigger Configuration"] += trigger_parameter_data

  def getBlockingTriggers(firing_trigger_id):
    df_dict["Trigger"] += "\n\n*** TRIGGER EXCEPTIONS ***\n\n".format()
    df_dict["Trigger Configuration"] += "\n*** TRIGGER EXCEPTIONS ***\n\n".format()
    getTriggers(firing_trigger_id)

  tag_count = 0
  for tag in tags:
    tag_type = tag['type']
    tag_paused = False
    if("paused" in tag and tag["paused"]):
      tag_paused = True
    if(tag_type == "ua" and not tag_paused):       # add tag_type == "ua" to get UA tags
      tag_count += 1
      tag_name = tag['name']
      df_dict["S.No"] = tag_count
      df_dict["Tag Name"] = tag_name
      df_dict["Tag Type"] = tag_type_map[tag_type] if (tag_type in tag_type_map) else tag_type
      firing_trigger_id = tag['firingTriggerId'] if ('firingTriggerId' in tag) else None
      blocking_trigger_id = tag['blockingTriggerId'] if ('blockingTriggerId' in tag) else None
      tag_parameters = tag.get('parameter')
      tag_parameter_data = ""
      if(tag_parameters):
        for tag_parameter in tag_parameters:
          if(tag_parameter["type"] == "BOOLEAN" or tag_parameter["type"] == "TEMPLATE"):
            tag_parameter_data += "{0} - {1}\n".format(tag_parameter['key'], tag_parameter['value'] if('value' in tag_parameter) else None)
          elif(tag_parameter["type"] == "LIST"):
            tag_parameter_data += "\n{0} - {1}".format(tag_parameter['key'], tag_parameter['type'])
            if('list' in tag_parameter):
              for list_item in tag_parameter["list"]:
                tag_parameter_data += "\n{0} - {1}".format(list_item['map'][0]['value'], list_item['map'][1]['value'])

      tag_parameter_data += "\n\nTag Priority - {0}".format(tag['priority']['value']) if('priority' in tag) else ''
      tag_parameter_data += "\n\nTag Firing Option - {0}".format(tag['tagFiringOption']) if('tagFiringOption' in tag) else ''
      tag_parameter_data += "\n\nSchedule Start - UTC {0}".format(str(datetime.datetime.fromtimestamp(int(tag['scheduleStartMs'])/1000.0))) if('scheduleStartMs' in tag) else ''
      tag_parameter_data += "\n\nSchedule End - UTC {0}".format(str(datetime.datetime.fromtimestamp(int(tag['scheduleEndMs'])/1000.0))) if('scheduleEndMs' in tag) else ''

      if('setupTag' in tag):
        for setup_tag in tag['setupTag']:
          tag_parameter_data += "\n\nFire a Tag before {0} fires:\n{1} - {2} Don't Fire".format(tag_name, setup_tag['tagName'], "Enable" if('stopOnSetupFailure' in setup_tag and setup_tag['stopOnSetupFailure']) else "Disable")

      if('teardownTag' in tag):
        for teardown_tag in tag['teardownTag']:
          tag_parameter_data += "\n\nFire a Tag after {0} fires:\n{1} - {2} Don't Fire".format(tag_name, teardown_tag['tagName'], "Enable" if('stopTeardownOnFailure' in teardown_tag and teardown_tag['stopTeardownOnFailure']) else "Disable")
      
      
      if('monitoringMetadataTagNameKey' in tag):
        tag_parameter_data += '\n\nTag Metadata:'
        tag_parameter_data += '\nMetadata Tag Name - {0}'.format(tag['monitoringMetadataTagNameKey']) if('monitoringMetadataTagNameKey' in tag) else ''
        for count, metadata in enumerate(tag['monitoringMetadata']['map']):
          tag_parameter_data += '\n{0}: {1} - {2}'.format(str(count+1), metadata['key'], metadata['value'])


      if('consentSettings' in tag):
        tag_parameter_data += '\n\nConsent Settings:'
        tag_parameter_data += '\nStatus - {0}'.format(tag['consentSettings']['consentStatus'])
        if(tag['consentSettings']['consentStatus'] == "NEEDED"):
          for count, list_item in enumerate(tag['consentSettings']['consentType']['list']):
            tag_parameter_data += '\n{0}: {1}'.format(str(count+1), list_item['value'])

      
      df_dict["Tag Configuration"] = tag_parameter_data

      getTriggers(firing_trigger_id) if (firing_trigger_id) else None

      getBlockingTriggers(blocking_trigger_id) if (blocking_trigger_id) else None
      
      df_dict = { key:str(value).strip() for key, value in df_dict.items()}
      if(tag_count == 1):
        df = pd.DataFrame(df_dict, index=[tag_count])
      else:
        df1 = pd.DataFrame(df_dict, index=[tag_count])
        df = df.append([df1])
        
      df_dict = {'S.No': "", 'Tag Name': "", 'Tag Type': "", 'Tag Configuration': "", 'Trigger': "", 'Trigger Configuration': ""}

  df['Container Info'] = "ID: {0}\nName: {1}\nType: {2}".format(container_id, container_name, container_type)

  df.set_index('S.No')

  if(result[account_id].empty):
    result[account_id] = df
  else:
    for i in range(5):
      result[account_id] = result[account_id].append(pd.DataFrame(df_dict_empty, index=[0]))
    result[account_id] = result[account_id].append(df)

  result[account_id].set_index('S.No')
  path = '/content/temp/{0} Internal - GTM Tags Extraction (Jul 2022).xlsx'.format(account)
  with open(path, 'w', encoding = 'utf-8-sig') as f:
    result[account_id].to_excel(path, sheet_name=account, index=False)
```
